10. Deal with mussing values - Filling

```python
revenue_mean = data_indexed['Revenue (Millions)'].mean()
print("The mean revenue is: ",revenue_mean)

#We can fill the null values with this mean revenue
data_indexed['Revenue (Millions)'].fillna(revenue_mean,inplace=True)
```

11. The apply() functions

```python
def rating_group(rating):
  if rating >= 7.5:
    return 'Good'
  elif rating >=6:
    return 'Average'
  else:
    return 'Bad'

# Lets apply this function on our movies data
# creating a new variable in the dataset to hold the raign category
df['Rating_Category'] = df['Rating'].apply(rating_group)

df[['Title','Director','Rating','Rating_Category']].head(5)
```
